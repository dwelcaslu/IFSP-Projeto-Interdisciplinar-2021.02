## Descrição do dataset e do problema a ser resolvido

Os abalones são moluscos, espécie de caracóis marinhos, que vivem preferencialmente em águas frias. Pertencem ao gênero Haliotis, da família Haliotidae. Sua concha é arredondada ou ovalada, achatadas, e nela vemos uma seqüência de furos (poros), variando de 2 a 9, conforme o tipo de abalone, seguidos por uma seqüência de pequenas elevações, sem furos. Sua cor externa varia de acordo com o seu habitat natural. Já na face interior, encontra-se uma belíssima e iridescente madrepérola, podendo ser negra, azul, verde, rosa, amarela, vermelha, branca, entre outras mais, que é muito apreciada para fabricação de jóias e adornos.

O objetivo do regressor projetado é prever a idade do abalone a partir de medições físicas. Dos dados originais, os exemplos com valores ausentes foram removidos (a maioria tendo o valor predito ausente) e os intervalos dos valores contínuos foram escalados para uso com uma ANN (dividindo por 200).

```python
# define the schema, corresponding to a line in the csv data file.
schema = StructType([
    StructField("sex", StringType(), nullable=True),
    StructField("length", FloatType(), nullable=True),
    StructField("diameter", FloatType(), nullable=True),
    StructField("height", FloatType(), nullable=True),
    StructField("whole_weight", FloatType(), nullable=True),
    StructField("shucked_weight", FloatType(), nullable=True),
    StructField("viscera_weight", FloatType(), nullable=True),
    StructField("shell_weight", FloatType(), nullable=True),
    StructField("rings", IntegerType(), nullable=True)]
)
```

## 5. Pré-processamento


```python
featureCols = ['length', 'diameter', 'height','whole_weight',
               'shucked_weight','viscera_weight','shell_weight']
# put features into a feature vector column
assembler = VectorAssembler(inputCols=featureCols, outputCol="features")
assembled_df = assembler.transform(abalone_df)
assembled_df
```

```python
assembled_df.show(truncate=False)
```

```python
# Initialize the `standardScaler`
standardScaler = StandardScaler(inputCol="features", outputCol="features_scaled")
```

```python
# Fit the DataFrame to the scaler
prep_df = standardScaler.fit(assembled_df).transform(assembled_df)
```

```python
# Inspect the result
prep_df.select("features", "features_scaled").show(10, truncate=False)
```

### 6. Construindo um modelo baseline

```python
# Initialize `lr`
lr = (LinearRegression(featuresCol='features_scaled', labelCol="rings", predictionCol='rings_pred',
                       maxIter=10, regParam=0.3, elasticNetParam=0.8, standardization=False))
```

```python
ybase_true = baseline_predictions["rings"].values
ybase_pred = baseline_predictions["rings_pred"].values
```

```python
def get_model_metrics(y_true, y_pred, model_name=''):
    model_metrics = {"mse": mean_squared_error(y_true, y_pred),
                     "mae": mean_absolute_error(y_true, y_pred),
                     "rmse": mean_squared_error(y_true, y_pred)**(1/2),
                     "r2": r2_score(y_true, y_pred),
                     "model": model_name,
                    }
    return model_metrics

baseline_metrics = get_model_metrics(ybase_true, ybase_pred, "baseline")
baseline_metrics
```

## 8. Modelos avançados

```python
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.regression import LinearRegression, RandomForestRegressor
from pyspark.ml import Pipeline
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
```

```python
numFolds = 5
```

```python
# Data Preparation
assembled_df = assembled_df.withColumnRenamed("rings", "label")

# Split the data into train and test sets
train_data, test_data = assembled_df.randomSplit([.8,.2], seed=rnd_seed)

standardScaler = StandardScaler(inputCol="features", outputCol="features_scaled")
```

```python
# Linear Regression
lr = LinearRegression(maxIter=30)

pipeline = Pipeline(stages=[standardScaler, lr])

# We use a ParamGridBuilder to construct a grid of parameters to search over.
# TrainValidationSplit will try all combinations of values and determine best model using
# the evaluator.
paramGrid = ParamGridBuilder()\
    .addGrid(lr.regParam, [0.5, 0.1, 0.05, 0.01, 0.005, 0.001, 0.0])\
    .addGrid(lr.fitIntercept, [False, True])\
    .addGrid(lr.elasticNetParam, [0.0, 0.01,0.05, 0.1, 0.5, 0.8, 1.0])\
    .build()

crossval = CrossValidator(estimator=pipeline,
                          estimatorParamMaps=paramGrid,
                          evaluator=RegressionEvaluator(),
                          numFolds=numFolds)

# Run cross-validation, and choose the best set of parameters.
LRModel = crossval.fit(train_data)
```

```python
# Random Forest
rf = RandomForestRegressor()
pipeline = Pipeline(stages=[standardScaler, rf])

# We use a ParamGridBuilder to construct a grid of parameters to search over.
# TrainValidationSplit will try all combinations of values and determine best model using
# the evaluator.
paramGrid = ParamGridBuilder()\
    .addGrid(rf.numTrees, [1,2,3,5,10,15,20])\
    .addGrid(rf.maxDepth, [1,2,3,5,10])\
    .addGrid(rf.minInstancesPerNode, [1,2,3,5,10])\
    .build()

crossval = CrossValidator(estimator=pipeline,
                          estimatorParamMaps=paramGrid,
                          evaluator=RegressionEvaluator(),
                          numFolds=numFolds)

# Run cross-validation, and choose the best set of parameters.
RFModel = crossval.fit(train_data)
```

## 8. Avaliação do(s) modelo(s) avançados(s)



```python
# Linear Regression model
lr_predictions = LRModel.bestModel.transform(test_data).select("features", "label", "prediction").toPandas()
lr_predictions
```

```python
ylr_true = lr_predictions["label"].values
ylr_pred = lr_predictions["prediction"].values

lr_metrics = get_model_metrics(ylr_true, ylr_pred, "LR model")
lr_metrics
```

```python
# Random Forest model
rf_predictions = RFModel.bestModel.transform(test_data).select("features", "label", "prediction").toPandas()
rf_predictions
```

```python
yrf_true = rf_predictions["label"].values
yrf_pred = rf_predictions["prediction"].values

rf_metrics = get_model_metrics(yrf_true, yrf_pred, "RF model")
rf_metrics
```

```python
def plot_model_comparison(y_true, y_pred_list, y_labels_list):
    x_array = np.arange(0, len(y_true), 1)
    
    plt.figure(figsize=(14, 10))
    plt.subplot(2,1,1)
    plt.title("sample predictions")
    plt.plot(x_array, ybase_true, color='gray',label='True values', alpha=1, linewidth=4,  zorder=2)
    for y_array,label in zip(y_pred_list, y_labels_list):
        plt.scatter(x_array, y_array, label=label, alpha=0.5, zorder=0)
    plt.xlim([x_array[0], x_array[-1]])
    plt.legend(loc='best')

    plt.subplot(2,1,2)
    plt.title("predictions distribution")
    sns.kdeplot(data=ybase_true, color='gray',label='True values', alpha=0.75, linewidth=4, zorder=2)
    for y_array,label in zip(y_pred_list, y_labels_list):
        plt.hist(y_array, label=label, alpha=0.75, bins=30, density=True, zorder=0)
    plt.legend(loc='best')

    plt.tight_layout()
    plt.show()
    
    
plot_model_comparison(ybase_true, [ybase_pred, ylr_pred, yrf_pred], ['Baseline', "LR model", "RF model"])
```
